导入库

```python
from tensorflow import keras
import tensorflow as tf
import numpy as np
import os
```

限制显存

```python
gpus = tf.config.list_physical_devices('GPU')
print(gpus)
if gpus:
    tf.config.experimental.set_virtual_device_configuration(gpus[0], [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=100)])
```

构建模型  

units 指定该层只有一个神经元  
input_shape 为输入值的形状，即一个 x

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

编译模型

optimizer 优化方式 - sgd 为随机梯度下降  
loss 损失函数 - mean_squared_error 为均方误差

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

准备数据

```python
x = np.array([-1,0,1,2,3,4], dtype=float)
y = np.array([-3,-1,1,3,5,7], dtype=float)
```

训练

epochs 为训练次数

```python
model.fit(x, y, epochs=1000)
```

测试预测

上面数据拟合的是
$y=2x-1$
，这里可以验证一下

```python
for i in range(5, 10):
    print('%g - %g' %(i, model.predict([i])))
```
