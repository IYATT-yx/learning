```python
import matplotlib.pyplot as plt
import numpy as np
```

列出所有风格

```python
plt.style.available
```

```python
def draw():
    x = np.linspace(-10, 10, 100)
    y = np.cos(x)
    plt.plot(x, y)
    plt.show()
```

```python
plt.style.use('dark_background')
draw()
```

```python
plt.style.use('seaborn-ticks')
draw()
```

```python
plt.style.use('ggplot')
draw()
```
